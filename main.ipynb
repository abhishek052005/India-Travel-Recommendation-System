# Popularity Based Recommender System

```python
rating_df = pd.DataFrame({'Name': df["Name"],
                              'Number of google review in lakhs': df["Number of google review in lakhs"],
                              'Google review rating': df["Google review rating"]
                              })
rating_df
```

```python
popularity_df = df[rating_df['Number of google review in lakhs']>=0.5].sort_values('Google review rating',ascending=False)
popularity_df
```

# Content based Filtering


```python
from sklearn.feature_extraction.text import TfidfVectorizer  # using for chech which word is more important in the document
from sklearn.metrics.pairwise import cosine_similarity





features = ['Zone', 'State', 'City', 'Type', 'Significance']
for feature in features:
    df[feature] = df[feature].fillna('').str.lower()



def create_soup(x):
    return f"{x['Zone']} {x['State']} {x['City']} {x['Type']} {x['Significance']}"

df['soup'] = df.apply(create_soup, axis=1)

#  Vectorize the Soup using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['soup'])

#  Compute the Similarity Matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)


def get_recommendations(place_name, cosine_sim=cosine_sim):
    # Get index of the place that matches the name
    try:
        idx = df[df['Name'].str.lower() == place_name.lower()].index[0]
    except IndexError:
        return "Place not found in dataset."


    sim_scores = list(enumerate(cosine_sim[idx]))


    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)


    sim_scores = sim_scores[1:6]


    place_indices = [i[0] for i in sim_scores]


    return df[['Name', 'City', 'Type', 'Google review rating']].iloc[place_indices]


print(get_recommendations("Taj Mahal"))
```

```python
import pickle

data_file = "india_travel_list.pkl"
similarity_file = "similarity_matrix.pkl"
popullarity_file = "popularity_df.pkl"

pickle.dump(df, open(data_file, 'wb'))

pickle.dump(cosine_sim, open(similarity_file, 'wb'))

pickle.dump(popularity_df, open(popullarity_file, 'wb'))

print("Model saved successfully!")
```
